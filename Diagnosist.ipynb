```python
cols_to_fill=['Glucose','BloodPressure','SkinThickness','BMI']
for col in cols_to_fill:
    df[col].replace(0,df[col].mean(),inplace=True)

df_sample=df.sample(18)
df_sample
```

```python
sns.countplot(data=df,x='Outcome')
plt.show()
```

```python
#counting number zero patient in each columns
nv=[]
#Identify columns with a minimum values of zero
for column_name in df.columns:
    if df[column_name].min()==0:
        nv.append(column_name)

print('columns with minimum values zero:',nv)

column=[]
value=[]

#Ignoring 'pregnancy' and 'toutconet' columns
for column_name in nv[1:-1]:
    zero_count=(df[column_name]==0).sum()
    column.append(column_name)
    value.append(zero_count)

dfr=pd.DataFrame({'column_name':column,'zero_count':value})
print(dfr)
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df,x='Age',bins=20,kde=True)
plt.show()
```

```python
#Finding the mode of the "Insulin" column
ins_mode=df['Insulin'].value_counts().index[0]

#Replacing 0 values in the 'Insulin' column with the mode
df['Insulin'].replace(0,ins_mode,inplace=True)

#Displaying a sample of 10 rows to verify the changes
print(df.sample(10))
```
